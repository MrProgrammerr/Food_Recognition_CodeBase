```python
import numpy as np
imgarray = []
for i in range(len(img)):
    arr = cv2.imread(str(img[i]))
    print(arr.shape)
    arr = cv2.resize(arr,(256,256))
    imgarray.append(arr)
imgarray = np.array(imgarray)
```

```python
imgarray.shape
```

```python
prd = model.predict(imgarray)
```

```python
pred = []
for i in prd:
    pred.append(i.argmax())
pred = np.array(pred)
```
